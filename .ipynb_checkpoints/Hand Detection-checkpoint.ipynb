```python
import uuid #unique identifier
import os
import time
```

```python
IMAGES_PATH = os.path.join('data','images') #/data/images
labels = ['ok', 'love', 'thumbsup']
number_imgs = 30
```

```python
cap = cv2.VideoCapture(0)
#loop thru labels
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    #loop thru img range
    for img_num in range(number_imgs):
        print('Collecting image for {}, image number{}'.format(label, img_num))
        #cam feed
        ret, frame = cap.read()
        #img path
        imgname = os.path.join(IMAGES_PATH,label+'.'+str(uuid.uuid1())+'.jpg')
        cv2.imwrite(imgname, frame)
        cv2.imshow('Image Collection', frame)
        #2 sec delay btw pics
        time.sleep(2)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp5/weights/last.pt', force_reload=True)
```

```python
%matplotlib inline 
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
